## Data binning
The positional data is sampled at a higher sampling rate than the spiking data, 39.0625 Hz and 20 kHz respectively. To be able to compare the res and the whl, we bin the spiking data into time windows which correspond to the time between position tracking samples. In other words we align the res with the whl.

Note: my spiking data is recorded at 24 kHz, but the _phy_to_clu_res_andrea.ipynb_ script downsamples the res file to 20 kHz.

To calculate the number of spiking samples that correspond to one position tracking sample, we divide the sampling rates:
$$
\frac{20,000 Hz}{39.0625 Hz} = 512
$$

Therefore, 512 spiking samples fit into one position tracking sample. We can use the formula period T = 1/F to calculate the time window:
$$
\frac{1}{39.0625 Hz} = 0.0256 seconds = 25.6 ms
$$

The function **bindata** initializes a 2d array with dimensions [total neurons+2][total time stamps sampled at 39.0625 Hz]. It iterates over the res file (ir is the index of the res element, r is the value) and assigns each spike time into the 25.6 ms time windows (bins it) by dividing the time stamp by 512 to find the corresponding bin.

```python
### bin spikes (aligned with whl - i.e. 25.6 ms time windows)
def bindata(clu,res):
    bindata = np.zeros([clu[0]+2, int(res[-1]/512)+1])
    for ir, r in enumerate(res):
        bindata[clu[ir+1], int(r/512)] += 1
    return bindata
```

### Computing the speed
Compute the speed from the whl as a simple moving average. This can be used for speed filtering.

```python
### compute speed - simple moving average
def comp_speed(whl): #compute speed from the whl - use it for filtering!
    temp = np.zeros(whl.shape[0])
    speed = np.zeros(whl.shape[0])
    for i in range(5,whl.shape[0]-5):
        if whl[i,0] > 0 and whl[i+1,0] > 0 and whl[i,1] > 0 and whl[i+1,1] > 0:
            temp[i] = np.sqrt((whl[i,0] - whl[i+1,0])**2 + (whl[i,1] - whl[i+1,1])**2)
        else:
            temp[i]=-1
    # smooth using a moving average
    for i in range(5,whl.shape[0]-5):
        t = temp[i-5:i+5]
        t = t[t>=0]
        if len(t)>0 and whl[i,0]>0:
            speed[i] = np.mean(t)
        else:
                speed[i]=np.nan
    return speed*39.0625
```

```python
speed = comp_speed(whl)
```

```python
### compute heading - also here simple moving average    
from scipy.stats import circmean
def heading(whl):
    temp = np.zeros(whl.shape[0])
    heading = np.zeros(whl.shape[0])
    for i in range(whl.shape[0]-1):
        if whl[i,0] > 0 and whl[i+1,0] > 0 and whl[i,1] > 0 and whl[i+1,1] > 0:
            temp[i] = np.arctan2(whl[i+1,1] - whl[i,1], whl[i+1,0] - whl[i,0]) + np.pi
        else:
            temp[i]=-1000

    for i in range(5,whl.shape[0]-5):
        t = temp[i-5:i+5]
        t = t[t>-1000]
        if len(t)>0 and whl[i,0]>0:
            heading[i] = circmean(t)
        else:
            heading[i] = np.nan
    return heading
```

```python
### Kernel density estimation of firing rate with triweight kernel
### this code is written for environemnts 150cm large and 3cm bins
### change parameters if you need more or less!

def sn(h,k): #squared norm
    y = h - k
    return y.T[0]**2 + y.T[1]**2

def K(x,z,sig): #triweight kernel
    
    nss = 9*sig**2 #9 sigma squared
    snp = 1 - sn(x,z) / nss #1 - squared norm points / 9 sigma squared
    snp[snp<0] = 0
    
    return 4 / (np.pi * nss) * snp**3
```

```python
def occmap(whl, speed=None,spf=None,sigma_ker=4.2,del_low_occ=True,): #occupancy map
    if spf:
        whl = np.array(whl[speed > spf,:])
    occ = np.zeros((50,50))
    for i in range(50): 
        for j in range(50):
            occ[i,j] = np.sum(K(np.array((1.5 + i*3, 1.5 + j*3)),whl,sigma_ker))
    if del_low_occ:  
	    occ[occ<np.max(occ)/1000]=0 #cut out bins with low coverage, < 0.1% of peak
    return occ
```
